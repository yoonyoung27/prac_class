###Git
'팀 개발을 위한 Git, GitHub 시작하기' 참고  
'(네이버 지식백과) 유닉스 리눅스 명령어 사전' https://terms.naver.com/list.naver?cid=59321&categoryId=59321  
'[GIT] 기본 환경 설정 및 작업 시작하기' https://www.tuwlab.com/ece/22210

* cd directory name : 해당 디렉터리로 이동

* git clone git repository address : 현 디렉터리에 깃 저장소 복사본 생성

* ls : 현 디렉터리의 파일 정보 출력

* git add file name : 로컬 저장소에 수정되거나 추가된 파일 임시 업데이트  
  -> git commit -m "explanation" : 수정 사항 기록  
  -> git push : 수정된 소스를 깃 저장소에 등록

* git pull : 현 로컬 저장소에 등록된 깃 저장소의 소스를 로컬 저장소에 업데이트

* mkdir directory name : 새 디렉터리 생성

* git status : 현재 작업 상태 출력
