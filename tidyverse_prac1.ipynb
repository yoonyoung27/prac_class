#①filter()
filter rows by their values

```
filter(flights, month == 1, day == 1) %>% head()
```

- __< head() >__ show first 6 objects  
- __< head(n) >__ show first n objects

```
flights %>% filter(month == 1 & day == 1 & dest %in% c('ORD','LAS')) %>% head()
```

#②arrange()
reorder the rows

```
arrange(flights, year, month, day) %>% head()
```

```
arrange(flights, desc(dep_delay)) %>% head()
```

- __< desc() >__ reorder in descending order

#③select()
select columns  by their names

```
select(flights, year, month, day) %>% head()
```

```
select(flights, year:day) %>% head()
```

- __< x:y >__ form x to y

```
select(flights, -(year:day)) %>% head()
```

- __< -x >__ except x

```
select(flights, starts_with("arr")) %>% head()
```

- __<starts_with("abc")>__ matches names that begin with “abc”.
- __<ends_with("xyz")>__ matches names that end with “xyz”.
- __<contains("ijk")>__ matches names that contain “ijk”.
- __<num_range("x", 1:3)>__ matches x1, x2 and x3.

```
select(flights, time_hour, air_time, everything()) %>% head()
```

- move the important columns in the first place

#④mutate()
create new variables with existing variables

```
flights %>% 
    select(year:day, ends_with("delay"), distance, air_time) %>% 
    mutate(gain = dep_delay - arr_delay,
              hours = air_time / 60,
              gain_per_hour = gain / hours) %>% head()
```

#⑤summarise()
grouped summaries

```
flights %>% group_by(year, month, day) %>% 
summarise(delay = mean(dep_delay, na.rm = TRUE)) %>% head()
```

- __<group_by()>__ grouping

#Ggplot
a package for data visualization

```
?mpg
```

#①geom_()

```
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point()
```

```
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy))
```

- mapping in ggplot() declare global mapping information

```
ggplot(data = mpg, mapping = aes(x = displ, y = hwy, color = class)) + geom_point()
```

- __< color(size, alpha, shape) = x >__ color(size, alpha, shape) represents x  

```
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point() + 
  geom_line()
```

- __<geom_line()>__ add graphical layers

#②facet_()
visualize multiple graphs

```
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + 
  facet_wrap(~ class, nrow = 2)
```

#③theme()
better visualization

```
ggplot(data = mpg) + 
  geom_point(mapping = aes(x = displ, y = hwy)) + theme_bw() +
  theme(axis.line= element_line(size=.8, colour = "black"),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        text = element_text(size = 25))
```
